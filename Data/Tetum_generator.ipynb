```python
import pandas as pd
from openai import OpenAI
API_KEY="x"
client = OpenAI(api_key=API_KEY)
 
INPUT_FILE = "species_en.xlsx"
OUTPUT_FILE = "species_tetum.xlsx"
 
TRANSLATE_COLUMNS = [
    "Common name",
    "Etymology",
    "Habitat",
    "Phenology",
    "Identification Characters",
    "Leaf type",
    "Fruit type",
    "Seed Germination",
    "Pest"
]
 
def translate_text(text):
    if pd.isna(text) or str(text).strip() == "":
        return ""
    prompt = f"Translate the following text to Tetum (Timor-Leste language). Keep meaning correct:\n\n{text}"
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[{"role": "user", "content": prompt}],
    )
    return response.choices[0].message.content.strip()
 
def translate_excel():
    print("Loading:", INPUT_FILE)
    df = pd.read_excel(INPUT_FILE)
    df_tet = df.copy()
    print("Translating English to Tetum...")
    for col in TRANSLATE_COLUMNS:
        if col in df_tet.columns:
            print("Translating column:", col)
            df_tet[col] = df_tet[col].apply(translate_text)
        else:
            print("Column not found:", col)
    df_tet["language"] = "tet"
    print("Saving Tetum Excel to:", OUTPUT_FILE)
    df_tet.to_excel(OUTPUT_FILE, index=False)
    print("Completed:", OUTPUT_FILE)
print("started")
if __name__ == "__main__":
    translate_excel()
```
